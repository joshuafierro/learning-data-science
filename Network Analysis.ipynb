# Network Analysis

<p>

A <i>network</i> is a body of connected data that's evaluated during graph analysis. Whereas a <i>graph</i> is the actual data visualization schematic that depicts the network data. 

</p>

## Use Cases:

<p>
    <ul>
        <li>Infrastructure system design</li>
        <li>financial risk management</li>
        <li>public health management</li>
    </ul>

</p>

<p>
    <b><i>Nodes</i></b> are vertices around which the graph is formed and edges are the lines that connect them. A <b><i>directed graph</i></b>, also known as a digraph, is a graph where there is a direction assigned to each edge that connects the nodes. And a <b><i>directed edge</i></b> is an edge feature that has been a assigned direction between those nodes.


</p>

<p>
<b>Undirected</b> graphs and those are graphs where all the edges are bidirectional and an undirected edge is just an edge that flows both ways between nodes. <b><i>Graph size</i></b> is the number of edges in a graph. <b><i>Graph</i></b> order is the number of vertices in a graph, and the term degree refers to the number of edges connected to a vertex. It's a measure of connectedness. 

</p>

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as ply
from pylab import rcParams
import seaborn as sns
import networkx as nx
```

```python
sns.set_style('whitegrid')
```

```python
G = nx.Graph()
G.add_nodes_from([2,3,4,5,6,7,8,9,12,15,16])
G.add_edges_from([(2,4),(2,6),(2,12),(2,16),(3,6),(3,9),(3,12),(3,15),(4,8),(4,12),(6,12),(8,16)])
nx.draw(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spring(G)
```

```python
#labeling

nx.draw_circular(G,node_color='bisque', with_labels=True)
```

```python
sum_stats = nx.info(G)
print(sum_stats)
```

```python
print(nx.degree(G))
```

## Graph Generators

```python
G = nx.complete_graph(25)
nx.draw(G,node_color='bisque',with_labels=True)
```

```python
G = nx.gnc_graph(7, seed=25)
nx.draw(G, node_color='bisque', with_labels=True)
```

```python
# ego graph
ego_G = nx.ego_graph(G, 3, radius=5)
nx.draw(G, node_color='bisque', with_labels=True)
```
